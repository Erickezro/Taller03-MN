```python
import numpy as np
import matplotlib.pyplot as plt

# Rango de valores para x
x = np.linspace(-4, 4, 400)

# Aproximaciones de Taylor para cos(x) con distintos trminos
cos_x_approx_2 = 1 - (x**2) / 2              # Aproximacin con 2 trminos
cos_x_approx_4 = 1 - (x**2) / 2 + (x**4) / 24  # Aproximacin con 4 trminos
cos_x_approx_6 = 1 - (x**2) / 2 + (x**4) / 24 - (x**6) / 720  # Aproximacin con 6 trminos

# Puntos x1 especficos para mostrar en el grfico
x1_points = [-3, -1, 0, 1, 3]
y_cos_x1 = np.cos(x1_points)  # Valores de cos(x) en los puntos x1

# Graficamos cada aproximacin
plt.figure(figsize=(10, 6))
plt.plot(x, np.cos(x), label='cos(x)', color='blue')  # Funcin original
plt.plot(x, cos_x_approx_2, label='2 trminos de Taylor', color='orange', linestyle='--')
plt.plot(x, cos_x_approx_4, label='4 trminos de Taylor', color='green', linestyle='--')
plt.plot(x, cos_x_approx_6, label='6 trminos de Taylor', color='red', linestyle='--')

# Configuraciones del grfico
plt.xlabel("x")
plt.ylabel("y")
plt.title("Aproximaciones de Taylor de la funcin cos(x)")
plt.legend()
plt.grid(True)
plt.ylim(-6,4)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as pltimport 

# Definir el rango de x (evitar valores cercanos a 1 para evitar divisin por cero)
x = np.linspace(-0.9, 0.9, 100)

# Funcin 1/(1 - x)
f = 1 / (1 - x)
                       
taylor_2 = 1 + x                     
taylor_3 = 1 + x + x**2              
taylor_4 = 1 + x + x**2 + x**3       
taylor_5 = 1 + x + x**2 + x**3 + x**4 


plt.figure(figsize=(15, 10))
plt.plot(x, f, label=r'$\frac{1}{1 - x}$', color='blue')
plt.plot(x, taylor_2, label='Segundo Orden', linestyle='--', color='green')
plt.plot(x, taylor_3, label='Tercer Orden', linestyle='--', color='brown')
plt.plot(x, taylor_4, label='Cuarto Orden', linestyle='--', color='orange')
plt.plot(x, taylor_5, label='Quinto Orden', linestyle='--', color='purple')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Serie de Taylor de la funcin 1/(1 - x)')
plt.legend()
plt.grid(True)  
plt.xlim(-0.9, 0.6)
plt.ylim(-0.1, 2)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Rango de valores para x
x = np.linspace(0.1, 2, 400)  

# Aproximaciones de Taylor para ln(x) con distintos trminos
ln_x_approx_1 = (x-1)
ln_x_approx_2 = (x-1) - 0.5*(x-1)**2
ln_x_approx_3 = (x-1) - 0.5*(x-1)**2 + (1/3)*(x-1)**3
ln_x_approx_4 = (x-1) - 0.5*(x-1)**2 + (1/3)*(x-1)**3 - (1/4)*(x-1)**4 
ln_x_approx_5 = (x-1) - 0.5*(x-1)**2 + (1/3)*(x-1)**3 - (1/4)*(x-1)**4 + (1/5)*(x-1)**5

# Puntos x1 
x1_points = [0.5, 1.2, 1.5, 1.8]
y_ln_x1 = np.log(x1_points)  

# Graficamos cada aproximacin
plt.figure(figsize=(10, 6))
plt.plot(x, np.log(x), label='ln(x)', color='blue')  
plt.plot(x, ln_x_approx_1, label='1 trmino de Taylor', color='orange', linestyle='--')
plt.plot(x, ln_x_approx_2, label='2 trminos de Taylor', color='green', linestyle='--')
plt.plot(x, ln_x_approx_3, label='3 trminos de Taylor', color='red', linestyle='--')
plt.plot(x, ln_x_approx_4, label='4 trminos de Taylor', color='skyblue', linestyle='--')
plt.plot(x, ln_x_approx_5, label='5 trminos de Taylor', color='brown', linestyle='--')

# Configuraciones del grfico
plt.xlabel("x")
plt.ylabel("y")
plt.title("Aproximaciones de Taylor de la funcin ln(x) cerca de x=1")
plt.legend()
plt.grid(True)
plt.xlim(0, 2)
plt.ylim(-2, 1)
plt.show()
```
